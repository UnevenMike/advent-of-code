```python
data = []
with open("inputs/1_1.txt", "r") as file:
    lines = file.readlines()
    data = [int(x) for x in lines]
    
data
```

```python
%%time
for i in range (len(data)):
    for j in range(i, len(data)):
        if data[i] + data[j] == 2020:
            print("Num 1: ", data[i])
            print("Num 2: ", data[j])
            print("Solution:", data[i] * data[j])
```
