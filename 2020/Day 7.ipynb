```python
data = open('inputs/7_1.txt', 'r').read().split("\n")
data
```

```python
rules = [ x.split(" bags contain ") for x in data]
```

```python
colors = [x[0] for x in rules]
colors.count('dotted olive bags')
```

```python
def build_entry(rule):
    if "no other bags" in rule[1]:
        return { rule[0]: None }
    insides = rule[1].replace(".", "").replace("bags","").replace("bag", "").split(", ")
    d = { rule[0]: [(t[0], t[2:].rstrip ()) for t in insides]}
    return d
```

```python
build_entry(rules[0])
```

```python
dic = {}
for r in rules:
    dic.update(build_entry(r))
```

```python
dic
```

```python
def search(start, data):
    acc = [start]
    if data[start] is None:
        return acc
    items = data[start]
    for x in items:
        d = search(x[1], data)
        acc = acc + d
    return acc
        
```

```python
search('light salmon', dic)
```

```python
%%time
groups = []
for x in dic.keys():
    groups.append(search(x, dic))
```

```python
count = 0
for x in groups:
    if 'shiny gold' in x:
        count+=1
```

```python
def calc(start, data):
    acc = 1
    print("searching: ", data[start])
    if data[start] is None:
        return acc
    items = data[start]
    for x in items:
        acc += int(x[0]) * calc(x[1], data)
    return acc
```

```python
calc('shiny gold', dic)
```
