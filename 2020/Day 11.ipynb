```python
class seat:
    def __init__(self, char, x, y):
        self.state = char
        self.pos = (x,y)
        
    def neighbors(self, seats, max_x, max_y):
        x = self.pos[0]
        y = self.pos[1]
        
        all_adj = []
        
        cur_x = x
        cur_y = y
        while (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y):
            cur_x = cur_x - 1
            cur_y = cur_y - 1
            if (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y) and seats[cur_x * max_y + cur_y].state != FLOOR:
                break
        all_adj.append((cur_x, cur_y))
            
        cur_x = x
        cur_y = y
        while (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y):
            cur_x = cur_x - 1
            cur_y = cur_y
            if (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y) and seats[cur_x * max_y + cur_y].state != FLOOR:
                break
        all_adj.append((cur_x, cur_y))
        
        cur_x = x
        cur_y = y
        while (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y):
            cur_x = cur_x - 1
            cur_y = cur_y + 1
            if (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y) and seats[cur_x * max_y + cur_y].state != FLOOR:
                break
        all_adj.append((cur_x, cur_y))
        
        cur_x = x
        cur_y = y
        while (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y):
            cur_x = cur_x
            cur_y = cur_y - 1
            if (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y< max_y) and seats[cur_x * max_y + cur_y].state != FLOOR:
                break
        all_adj.append((cur_x, cur_y))
        
        cur_x = x
        cur_y = y
        while (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y):
            cur_x = cur_x 
            cur_y = cur_y + 1
            if (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y) and seats[cur_x * max_y + cur_y].state != FLOOR:
                break
        all_adj.append((cur_x, cur_y))
        
        cur_x = x
        cur_y = y
        while (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y):
            cur_x = cur_x + 1
            cur_y = cur_y - 1
            if (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y) and seats[cur_x * max_y + cur_y].state != FLOOR:
                break
        all_adj.append((cur_x, cur_y))
        
        cur_x = x
        cur_y = y
        while (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y):
            cur_x = cur_x + 1
            cur_y = cur_y
            if (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y) and seats[cur_x * max_y + cur_y].state != FLOOR:
                break
        all_adj.append((cur_x, cur_y))
        
        cur_x = x
        cur_y = y
        while (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y):
            cur_x = cur_x + 1
            cur_y = cur_y + 1
            if (cur_x >= 0 and cur_x < max_x and cur_y >= 0 and cur_y < max_y) and seats[cur_x * max_y + cur_y].state != FLOOR:
                break
        all_adj.append((cur_x, cur_y))
        
        
        adj = [pos for pos in all_adj if (pos[0] >= 0 and pos[0] < max_x and pos[1] >= 0 and pos[1] < max_y)]
        indexs = [ max_y * pos[0] + pos[1] for pos in adj ]
        if (8191 in indexs): 
            print(self.pos, all_adj, adj)
        return indexs
                            

class lobby:
    def __init__(self, x, y, data):
        self.x = x
        self.y = y
        self.seats = []
        for i in range(x):
            for j in range(y):
                self.seats.append(seat(data[i][j], i, j))
                
    def inc(self):
        new_seats = []
        for i in range(self.x):
            for j in range(self.y):
                chair = self.seats[self.y * i + j]
                
                neighbors = chair.neighbors(self.seats, self.x, self.y)
                states = [s.state for s in [self.seats[ind] for ind in neighbors]]
                s = "".join(states)
                full = s.count(CLOSED);
                    
                if chair.state == FLOOR:
                    new_seats.append(seat(FLOOR, i, j))
                elif chair.state == OPEN:
                    new_seats.append(seat(CLOSED if full == 0 else OPEN ,i,j))
                elif chair.state == CLOSED:
                    new_seats.append(seat(OPEN if full >= 5 else CLOSED ,i,j))
        
        isSame = True;
        
        for i in range(self.x):
            for j in range(self.y):
                old_state = self.seats[self.y * i + j].state
                new_state = new_seats[self.y * i + j].state
                if old_state != new_state:
                    isSame = False
                    
        self.seats = new_seats
        return isSame
                    
```
