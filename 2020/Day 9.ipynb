```python
input_data = open('inputs/9_1.txt', 'r').read().split("\n")
input_data = [int(x) for x in input_data]
input_data
```

```python
preamble = 25
goal = 25918798;
#goal = 127
```

```python
for index in range(preamble, len(input_data)):
    yes = 0
    for i in range (index-preamble, index):
        for j in range(index-preamble, index):
            if (i != j) and (input_data[i] + input_data[j] == input_data[index]):
                yes = 1
    if yes == 0:
        print(input_data[index], yes)
```

```python
for i in range(len(input_data)):
    count = 0;
    index = i + 1
    while index < len(input_data) and count <= goal:
        count += input_data[index]
        if (count == goal):
            print(i, index, input_data[i], input_data[index])
        index += 1
        
```
