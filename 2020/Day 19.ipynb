```python
input_data = open('inputs/19.txt', 'r').read().split('\n\n')
input_data = [x.split('\n') for x in input_data]
```

```python
rules = {}
for r in input_data[0]:
    rule = r.split(": ")
    rules.update({int(rule[0]): rule[1].split(' | ') if '"' not in rule[1] else rule[1].replace('"', '')} )
```

```python
rules
```

```python
def is_left_valid(string, rule):
    rule_def = rules[rule]
    if(string == ''):
        return (False, 0)
    
    true_42_count = 0;
    true_31_count = 0;
    if isinstance(rule_def, list):
        valid_checks = []
        for or_rule in rule_def:
            rs = [int(s) for s in or_rule.split(" ")]
            pass_single_rule_checks = True
            offset_count = 0
               
            for i in range(len(rs)):                        
                if (8 == rs[i] or 11 == rs[i]):
                    keep_looping = True
                    while keep_looping:
                        res = is_left_valid(string[offset_count:], 42)

                        if res[0] == False:
                            offset_count += res[1]
                            keep_looping = False
                        else:
                            pass_single_rule_checks = res[0]
                            offset_count += res[1]
                            true_42_count += 1
                    
                    if rs[i] == 11:
                        keep_looping_31 = True
                        while keep_looping_31:
                            res = is_left_valid(string[offset_count:], 31)
                            if res[0] == False:
                                offset_count += res[1]
                                keep_looping_31 = False
                            else:
                                pass_single_rule_checks = res[0]
                                offset_count += res[1]
                                true_31_count += 1
                        
                    pass_single_rule_checks = (rs[i] == 8 and true_42_count > 0) or (rs[i] == 11 and (true_42_count - 1) >= true_31_count and true_31_count > 0) 
                else:
                    res = is_left_valid(string[offset_count:], rs[i])
                    pass_single_rule_checks = pass_single_rule_checks and  res[0]
                    offset_count += res[1]
            valid_checks.append((pass_single_rule_checks, offset_count, rs, string))
        
        for x in valid_checks:
            if x[0] == True:
                return(x)
        return (False, 0)
        
    else:
        return (string[0] == rule_def, 1)
    
def is_valid(string, rule):
    left_valid = is_left_valid(string, rule)
#     print(left_valid)
    return left_valid[0] and left_valid[1] == len(string)
```
