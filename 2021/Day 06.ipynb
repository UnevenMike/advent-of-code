```python
class Fish:
    def __init__(self, n, is_new = False):
        self.n = n
        self.is_new = is_new
        self.is_dead = False
        if (is_new):
            self.n += 2
            
    def inc(self):
        if self.n == 0:
            self.n = 6
            return Fish(6, True)
        else:
            self.n += -1
            return None
        
    def kill(self):
        self.is_dead = True
        
    def __str__(self):
        return str(self.n)
```

# Dynamic Programming

Build 'knowledge base' of how 1 fish grows over time:

1. Incrementing our way from 1 to 257 days, we want to experiment population growth starting with a single Fish(1) (i.e. how many fish will be alive after that number of days).

2. At the end of that number of days, we store the number of total fish in a dictionary with keys being the number of days.

3. While computing the total number of fish for a given x days, check if there are any fish alive with an 'n-value' of 1 and if x is in our dictionary. Dill these fish and just add the known value of the fishes lineage growth to a counter that will be included in the dictionary update.

4. Removing these fish from the population during experiments prevents the population from growing exponentionally, thus requiring an incredible amount of computation to understand the population growth.


```python
known_dict = {}
for remaining_days in range(1, 257):    
    fishes = [Fish(1)]
    new_fishes = []
    count = 0
    for num_days in range(remaining_days, 0, -1):
        for fish in fishes:
            if fish.n == 1 and num_days in known_dict:
                fish.kill()
                count += known_dict[num_days]
        fishes = list(filter(lambda f: not f.is_dead, fishes))
        for fish in fishes:
            f = fish.inc()
            if f:
                new_fishes.append(f)
        fishes = fishes + new_fishes
        new_fishes = []
    known_dict.update({remaining_days: len(fishes) + count})
```
