```python
class BingoBoard:
    def __init__(self, nums):
        self.board = list(map(lambda row: [(int(x), False) for x in row.split(' ') if x], nums))
        self.is_done = False

    def call_number(self, num):
        for i in range(0, 5):
            for j in range(0, 5):
                if self.board[i][j][0] == num:
                    self.board[i][j] = (self.board[i][j][0], True)
                    
    def is_bingo(self):
        for row in self.board:
            if all([a[1] for a in row]):
                self.is_done = True
                return True
        for i in range(0, 5):
            if all([row[i][1] for row in self.board]):
                self.is_done = True
                return True
        self.is_done = False
        return False
    
    def score(self, num):
        score = 0
        for i in range(0, 5):
            for j in range(0, 5):
                if not self.board[i][j][1]:
                    score += self.board[i][j][0]
        score = score * num
        return score
```

```python
data = []
with open("input_04.txt", "r") as file:
    lines = file.readlines()
    data = [x.replace('\n', '') for x in lines]
    
calls = [int(x) for x in data[0].split(",")]
data = data[2:]
```

```python
data
```

```python
boards = []
board = []
for line in data:
    if line == '':
        boards.append(board)
        board = []
    else:
        board.append(line)
boards.append(board)
```

```python
bbs = [BingoBoard(b) for b in boards]
bbs
```

```python
for call in calls:
    for bb in bbs:
        bb.call_number(call)
    for bb in bbs:
        if (not bb.is_done) and bb.is_bingo():
            print(bb.score(call))

        
```
