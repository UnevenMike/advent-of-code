```python
data = []
with open("input_01_1.txt", "r") as file:
    lines = file.readlines()
    data = [int(x) for x in lines]
data
```

```python
offsets = [ d - data[i] for i, d  in enumerate(data[1:])]
```

```python
bool_offsets = [ 1 if o > 0 else 0 for o in offsets]
```

```python
sum(bool_offsets)
```

```python
windows = [ d + data[i] + data[i+1] for i,d in enumerate(data[2:])]
```

```python
windows_offsets = [ d - windows[i] for i, d  in enumerate(windows[1:])]
```

```python
windows_bool_offsets = [ 1 if o > 0 else 0 for o in windows_offsets]
```

```python
sum(windows_bool_offsets)
```
