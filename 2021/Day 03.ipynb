```python
data = []
with open("input_03.txt", "r") as file:
    lines = file.readlines()
    data = [x.replace("\n", "") for x in lines]
    
data
```

```python
length = len(data[0])
```

```python
length
```

```python
gamma = []
epsilon = []
for i in range(0, length):
    bits = [line[i] for line in data]
    if bits.count('0') > bits.count('1'):
        gamma.append(0)
        epsilon.append(1)
    elif bits.count('1') > bits.count('0'):
        gamma.append(1)
        epsilon.append(0)
    else:
        print('tie')
```

```python
gamma, epsilon
```

```python
g = int("".join(str(x) for x in gamma), 2)
e = int("".join(str(x) for x in epsilon), 2)
```

```python
g, e, g*e
```

```python
oxygen_data = data.copy()
for i in range(0, length):
    bits = [line[i] for line in oxygen_data]
    char = ''
    if bits.count('0') > bits.count('1'):
        char = '0'
    elif bits.count('0') < bits.count('1'):
        char = '1'
    else:
        char = '1'
    oxygen_data = list(filter(lambda x: x[i] == char, oxygen_data))
    if len(oxygen_data) == 1:
        print(i)
        break;
```
