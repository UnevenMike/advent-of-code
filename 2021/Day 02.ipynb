```python
data = []
with open("input_02_1.txt", "r") as file:
    lines = file.readlines()
    data = [(x.split(" ")[0], int(x.split(" ")[1])) for x in lines]
data
```

```python
horizontal = 0
depth = 0
```

```python
for x in data:
    if x[0] == 'forward':
        horizontal += x[1]
    elif x[0] == 'down':
        depth += x[1]
    elif x[0] == 'up':
        depth -= x[1]
```

```python
horizontal, depth, horizontal*depth
```

```python
horizontal = 0
depth = 0
aim = 0
```

```python
for x in data:
    if x[0] == 'forward':
        horizontal += x[1]
        depth += (aim * x[1])
    elif x[0] == 'down':
        aim += x[1]
    elif x[0] == 'up':
        aim -= x[1]
```

```python
horizontal, depth, aim, horizontal*depth
```
