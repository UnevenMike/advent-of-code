```python
def increment(number):
    index = 5
    numbers[index] = (numbers[index] + 1) % 10
    while index > 0 and numbers[index] == 0:
        index -= 1
        numbers[index] = (numbers[index] + 1) % 10
        
    return numbers   
```

```python
def check_valid(number):
    return increasing_check(number) and double_check(number)

def increasing_check(number):
    value = number[0]
    for i in range(1, 6):
        if number[i] < value:
            return False
        value = number[i]
    return True

def double_check(number):
    if number[0] == number[1] and number[1] != number[2]:
        return True
    if number[0] != number[1] and number[1] == number[2] and number[2] != number[3]:
        return True
    if number[1] != number[2] and number[2] == number[3] and number[3] != number[4]:
        return True
    if number[2] != number[3] and number[3] == number[4] and number[4] != number[5]:
        return True
    if number[3] != number[4] and number[4] == number[5]:
        return True

    return False
```

```python
# numbers = [1, 5, 6, 2, 1, 8]
numbers = [1, 1, 1, 1, 1, 1]
```

```python
print(increment(numbers))
print(check_valid(numbers))
```

```python
count = 0
numbers = [1, 5, 6, 2, 1, 8]
while numbers != [6, 5, 2, 5, 2, 7]:
    if check_valid(numbers):
        count += 1
        print(numbers)
    increment(numbers)
    
count
```
