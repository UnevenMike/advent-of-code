```python
import operator
import sys
```

```python
def read_file_to_list(filename):
    with open(filename) as f:
        content = f.readlines()
    content = [c.rstrip() for c in content]
    content = [c.split(",") for c in content]
    content = [c for c in content]
    return content
```

```python
def generate_points(path):
    x = 0
    y = 0
    distance = 0
    points = set()
    triplets = {}
    for line in path:
        direction = line[0]
        length = int(line[1:])
        for i in range (1, length + 1):
            if direction == 'R':
                x = x + 1
            elif direction == 'L':
                x = x - 1
            elif direction == 'U':
                y = y + 1
            elif direction == 'D':
                y = y - 1
            distance = distance + 1
            
            if not (x,y) in points:
                points.add((x,y))
                triplets[(x,y)] = distance
                
    return (points, triplets)
```

```python
def min_man_distance(dict1, dict2, crosses):
    min = sys.maxsize
    for c in crosses:
        distance = dict1[c] + dict2[c]
        if distance < min:
            min = distance
            
    return min
```
