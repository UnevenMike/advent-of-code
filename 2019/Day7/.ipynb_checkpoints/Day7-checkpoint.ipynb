```python
from itertools import permutations 
```

```python
def read_file_to_list(filename):
    with open(filename) as f:
        content = f.readlines()
    content = content[0].split(",")
    content = [int(c) for c in content]
    return content
```

```python
def get_value(tape, ins_ptr, param):
#     print("get", tape, ins_ptr, param)
    v = 0
    if param: #immediate mode
        v = tape[ins_ptr]
    else:
        v = tape[tape[ins_ptr]]
    return int(v)

def write_value(tape, ins_ptr, value, param):
#     print("write", tape, ins_ptr, param)
    if param: #immediate mode
        tape[ins_ptr] = value
    else:
        tape[tape[ins_ptr]] = value
#     print(tape)
    return tape
```

```python
def compute(tape, instruction_ptr, inputs, num_inputs, output):
    index = instruction_ptr
    value = 0
    inc = 0
    input_counter = 0
    
    # Get First instruction 
    instruction = tape[index]
    op_code = instruction % 100
    param = [None, instruction % 1000 // 100, instruction % 10000 // 1000, instruction % 100000 // 10000 ]
    
    while not op_code == 99:
        if op_code == 1:
            value = get_value(tape,index+1,param[1]) + get_value(tape, index+2, param[2])
            tape = write_value(tape, index+3, value, param[3])
            inc = 4
            
        elif op_code == 2:
            value = get_value(tape,index+1,param[1]) * get_value(tape, index+2, param[2])
            tape = write_value(tape, index+3, value, param[3])
            inc = 4
            
        elif op_code == 3:
            value = None
            if input_counter < num_inputs:
                value = inputs[input_counter]
                input_counter += 1
            else:
                # Need to get input later - save tape and inst_ptr
                return index
                # value = int(input())
                
            tape = write_value(tape, index+1, value, param[1])
            inc = 2
            
        elif op_code == 4:
            value = get_value(tape, index+1, param[1])
            output[0] = value
            inc = 2
            
        elif op_code == 5:
            if get_value(tape, index+1, param[1]) != 0:
                index = get_value(tape, index+2, param[2])
                inc = 0
            else:
                inc = 3
                
        elif op_code == 6:
            if get_value(tape, index+1, param[1]) == 0:
                index = get_value(tape, index+2, param[2])
                inc = 0
            else:
                inc = 3
                
        elif op_code == 7:
            value = 1 if get_value(tape, index+1, param[1]) < get_value(tape, index+2, param[2]) else 0
            tape = write_value(tape, index+3, value, param[3])
            inc = 4
        
        elif op_code == 8:
            value = 1 if get_value(tape, index+1, param[1]) == get_value(tape, index+2, param[2]) else 0
            tape = write_value(tape, index+3, value, param[3])
            inc = 4
        
        # Get next instruction and parameters
        index = index + inc            
        instruction = tape[index]
        op_code = instruction % 100
        param = [None, instruction % 1000 // 100, instruction % 10000 // 1000, instruction % 100000 // 10000 ]
    return None
```

```python
max_signal = 0
max_perm = None
instructions = read_file_to_list("input.txt")
phases = [5,6,7,8,9]
perms = permutations(phases)

for p in perms:
    # Set initial state for each computer
    tapes = [instructions.copy(), instructions.copy(), instructions.copy(), instructions.copy(), instructions.copy()]
    inputs = [[p[0], 0], [p[1]], [p[2]], [p[3]], [p[4]]]
    outputs = [[0], [0], [0], [0], [0]]
    indexs = [0, 0, 0, 0, 0]

    # loop throuhg computers until the last computer ends
    comp_index = 0
    while indexs[4] != None: 
        indexs[comp_index] = compute(tapes[comp_index], 
                                     indexs[comp_index], 
                                     inputs[comp_index], 
                                     len(inputs[comp_index]), 
                                     outputs[comp_index])
        inputs[comp_index] = []
        value = outputs[comp_index][0]
        comp_index = (comp_index + 1) % 5
        inputs[comp_index].append(value)

    if outputs[4][0] > max_signal:
        max_signal = outputs[4][0]
        max_perm = p
        
print(max_signal, max_perm)
```
